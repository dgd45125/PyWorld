```python
'''
The Actor class is the parent of the Knight, Rogue,
and Orc classes. Actors have an x,y position on the game map,
a .png sprite, attack, defense, hp, and stamina stats, a name,
gold and kill counts, and a priority.
'''
class Actor:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.png = None
        self.attack_power = None
        self.defense= 0
        self.hp = 8
        self.stamina =128
        self.name = None
        self.gold = 0
        self.kills = 0
        self.priority = None
    '''
    The draw method blits the actor sprite, the actor's helth bar
    and if not an orc, the actor's stamina bar
    '''
    def draw(self):
        #draw the actor
        MAIN_SURFACE.blit(self.png, (self.x*CELL_WIDTH,self.y*CELL_HEIGHT))
        #draw health
        if self.hp == 8:
            MAIN_SURFACE.blit(FULL_HEALTH, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-12))
        elif self.hp == 7:
            MAIN_SURFACE.blit(SEVEN_EIGHT_HEALTH, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-12))
        elif self.hp == 6:
            MAIN_SURFACE.blit(THREE_QUARTER_HEALTH, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-12))
        elif self.hp == 5:
            MAIN_SURFACE.blit(FIVE_EIGHT_HEALTH, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-12))
        elif self.hp ==4:
            MAIN_SURFACE.blit(HALF_HEALTH, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-12))
        elif self.hp ==3:
            MAIN_SURFACE.blit(THREE_EIGHT_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-12))
        elif self.hp ==2:
            MAIN_SURFACE.blit(QUARTER_HEALTH, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-12))
        elif self.hp ==1:
            MAIN_SURFACE.blit(ONE_EIGHT_HEALTH, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-12))
        
        #no stamina for orcs
        if (isinstance(self,Orc)):
            return
        else:
            #draw stamina
            if (self.stamina<=128 and self.stamina>=113):
                MAIN_SURFACE.blit(FULL_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-22))
            elif (self.stamina<=112 and self.stamina>=97):
                MAIN_SURFACE.blit(SEVEN_EIGHT_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-22))
            elif (self.stamina<=96 and self.stamina>=81):
                MAIN_SURFACE.blit(THREE_QUARTER_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-22))
            elif (self.stamina<=80 and self.stamina>=65):
                MAIN_SURFACE.blit(FIVE_EIGHT_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-22))
            elif (self.stamina<=64 and self.stamina>=49):
                MAIN_SURFACE.blit(HALF_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-22))
            elif (self.stamina<=48 and self.stamina>=33):
                MAIN_SURFACE.blit(THREE_EIGHT_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-22))
            elif (self.stamina<=32 and self.stamina>=17):
                MAIN_SURFACE.blit(QUARTER_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-22))
            elif (self.stamina<=16 and self.stamina>=1):
                MAIN_SURFACE.blit(ONE_EIGHT_STAMINA, (self.x*CELL_WIDTH-8,self.y*CELL_HEIGHT-22))  
    '''
    The move method, at its core, changes the map location of the actor
    along the x or y axis. It also incorporates picking up items and determining
    whether or not the actor should attack another actor. When items are picked up,
    they are simply moved to (-1,-1) where they cannot be seen. It also handles killing
    off actors due to stamina dropping below zero. They are simply moved to (-1,-1)
    '''                
    def move(self, deltaX, deltaY):
        #if item, do item action
        for item in GAME_ITEMS:
            #orcs dont get to pick up items and dead heros don't pick up items
            if (item.x==(self.x+deltaX)and item.y==(self.y+deltaY) and not(isinstance(self,Orc)) and not(self.x==-1 and self.y==-1)):
                if (isinstance(item,Gold)):
                    #hide item off screen
                    item.x = -1
                    item.y=-1
                    self.gold += 1
                    GAME_MESSAGES.append("The "+self.name+" has found gold")
                    GAME_MESSAGES.append("The "+self.name+" has collected "+str(self.gold)+" gold")
                if (isinstance(item,Fruit)):
                    #hide item off screen
                    item.x = -1
                    item.y = -1
                    GAME_MESSAGES.append("The "+self.name+" has found and eaten a fruit")
                    self.stamina +=64
                    if (self.stamina >128):
                        self.stamina = 128
                    GAME_MESSAGES.append("The "+self.name+"'s stamina is "+str(self.stamina)+"/"+str(128)) 
                if (isinstance(item,Heart)):
                    #hide item off screen 
                    item.x = -1
                    item.y = -1
                    self.hp += 4
                    GAME_MESSAGES.append("The "+self.name+" has found a heart")
                    if (self.hp > 8):
                        self.hp = 8
                    GAME_MESSAGES.append("The "+self.name+"'s health is "+str(self.hp)+"/"+str(8))
            
        #ensure actor cannot walk through actors and trees
        if not(Is_Occupied_Actor(self.x + deltaX,self.y + deltaY)) and not(Is_Tree(self.x + deltaX,self.y + deltaY)):
            self.x += deltaX
            self.y += deltaY
            #decrement stamina 1 point for each step taken
            self.stamina-=1
            
            #if a hero (not orc)'s stamina drops below 1, they are dead and will
            #be placed in -1,-1 with the other dead things and no longer receive a ply in the game loop
            if ((isinstance(self,Knight) and self.stamina <=0) or (isinstance(self,Rogue) and self.stamina <=0)):
                GAME_MESSAGES.append("The "+self.name+" has died of exhaustion")
                self.x=-1
                self.y=-1
            
        #if enemy actor, attack
        elif (Is_Occupied_Actor(self.x + deltaX,self.y + deltaY)):
            #if actor doesn't move, make sure it doesn't attack self
            if (deltaX==0 and deltaY==0):
                return
            for actor in GAME_ACTORS:
                #nobody attacks themselves
                if (actor.x ==(self.x+deltaX)and actor.y ==(self.y+deltaY)):
                    #rogue should not attack knight
                    if (isinstance(self,Rogue) and isinstance(actor,Knight)):
                        return
                    #knight should not attack rogue
                    if (isinstance(self,Knight) and isinstance(actor,Rogue)):
                        return
                    #orc should not attack orc
                    if (isinstance(self,Orc) and isinstance(actor,Orc)):
                        return
                    self.attack(actor)
                    self.stamina-=4
                    
                    #if a hero (not orc)'s stamina drops below 1, they are dead and will
                    #be placed in -1,-1 with the other dead things and no longer receive a ply in the game loop
                    if ((isinstance(self,Knight) and self.stamina <=0) or (isinstance(self,Rogue) and self.stamina <=0)):
                        GAME_MESSAGES.append("The "+self.name+" has died of exhaustion")
                        self.x=-1
                        self.y=-1
                    
        else:
            return
    '''
    The attack method reduces the defender's hp by
    (the attacker's attack stat minus the defender's defense stat).
    It also handles dead orcs dropping loot and killing off actors
    whose hp drops below zero. They are simple moved to (-1,-1)
    '''        
    def attack(self, actor):
        #reduce enemy hp by self.attack_power - actor.defense
        actor.hp = actor.hp -self.attack_power + actor.defense
        GAME_MESSAGES.append("The "+self.name+" attacks the "+actor.name+" for "+str(self.attack_power-actor.defense)+" damage")
        
        if (actor.hp <=0):
            #if dead orc, drop some loot if it is available
            if(isinstance(actor,Orc)):
                existsItem = False
                #check that an item is available
                for item in GAME_ITEMS:
                    if (item.x==-1 and item.y==-1):
                        existsItem = True
                goodSelect = False
                if (existsItem):
                    #pull a random item but make sure it is not in map
                    randItem = random.choice(GAME_ITEMS)
                    while not(goodSelect):
                        randItem = random.choice(GAME_ITEMS)
                        if(randItem.x==-1 and randItem.y==-1):
                            goodSelect = True
                    #replace orc corpse with random unused item
                    randItem.x = actor.x
                    randItem.y = actor.y
            
            GAME_MESSAGES.append("The "+actor.name+" is dead")
            self.kills +=1
            GAME_MESSAGES.append("The "+self.name+" has killed "+str(self.kills)+" orcs")
            #place actor corpse off screen
            actor.x = -1
            actor.y = -1
            
        else:
            GAME_MESSAGES.append("The "+actor.name+"'s health is "+str(actor.hp)+"/"+str(8))
        return
    '''
    Filler method that is defined for each child class
    '''       
    def ply():
        return     
```

### Actor: Knight

```python
'''
The Knight actor is a child of the actor class. It has an 
attack stat of 4 and a defense stat of 1. The Knight's 
primary priority is hunting down and killing orcs. If it's
hp or stamina drops below half, it will change it's priority
to go find health or food.
'''
class Knight(Actor):
    def __init__(self,x,y):
        Actor.__init__(self, x, y)
        self.png = KNIGHT
        self.attack_power = 4
        self.name = "Knight"
        self.defense = 1

    def ply(self):

        #priority is initially set to killOrc each ply
        self.priority = 'killOrc'
        #if stamina is less than half, that becomes priority
        if (self.stamina <=64):
            self.priority = 'stamina'
        #if hp is less than half, that becomes priority
        if (self.hp<=4):
            self.priority = 'hp'
    
        #regain hp
        if (self.priority == 'hp'):
            #check that there exists Heart in play
            existsHeart = False
            for item in GAME_ITEMS:
                if (isinstance(item,Heart)and not (item.x==-1)):
                    existsHeart = True
            #if no heart in play, go find fruit
            if (existsHeart == False):
                self.priority = 'stamina'
                
            if (existsHeart == True):
                #get x and y of nearest Heart
                heartDistance = 1000
                heartX = 1000
                heartY = 1000
                for item in GAME_ITEMS:
                    #get items that are Heart and not out of play
                    if (isinstance(item, Heart) and not(item.x==-1)):
                        distance = abs(self.x-item.x)+abs(self.y-item.y)
                        if (distance < heartDistance):
                            heartDistance = distance
                            heartX = item.x
                            heartY = item.y

                #move towards closest heart and navigate around trees
                if (self.x > heartX):
                    #pass on right side
                    if (Is_Tree(self.x-1,self.y-1)and not (Is_Tree(self.x,self.y-1))):
                        self.move(0,-1)
                    elif (Is_Tree(self.x-1,self.y)and not Is_Tree(self.x-1,self.y-1)):
                        self.move(0,-1)
                    else:
                        self.move(-1,0)
                        
                elif(self.x < heartX):
                    #pass on left side
                    if (Is_Tree(self.x+1,self.y+1)and not (Is_Tree(self.x,self.y+1))):
                        self.move(0,1)
                    elif (Is_Tree(self.x+1,self.y)and not Is_Tree(self.x+1,self.y+1)):
                        self.move(0,1)
                    else:
                        self.move(1,0)
                        
                elif(self.y < heartY):
                    if (Is_Tree(self.x,self.y+1)):
                        self.move(-1,0)
                    else:
                        self.move(0,1)
                        
                #(self.y > heartY)       
                else:
                    if (Is_Tree(self.x,self.y-1)):
                        self.move(1,0)
                    else:
                        self.move(0,-1)    
    
        #regain stamina      
        if (self.priority == 'stamina'):

            #check that there exists Fruit in play
            existsFruit = False
            for item in GAME_ITEMS:
                if (isinstance(item,Fruit) and not (item.x==-1)):
                    existsFruit = True
            #if no fruit, find gold
            if (existsFruit == False):
                self.priority = 'gold'
            if (existsFruit == True):
                #get x and y of nearest fruit
                fruitDistance = 1000
                fruitX = 1000
                fruitY = 1000
                for item in GAME_ITEMS:
                    #get items that are fruit and not out of play
                    if (isinstance(item, Fruit) and not(item.x==-1)):
                        distance = abs(self.x-item.x)+abs(self.y-item.y)
                        if (distance < fruitDistance):
                            fruitDistance = distance
                            fruitX = item.x
                            fruitY = item.y
                #move towards closest fruit and navigate around trees
                if (self.x > fruitX):
                    #pass on right side
                    if (Is_Tree(self.x-1,self.y-1)and not (Is_Tree(self.x,self.y-1))):
                        self.move(0,-1)
                    elif (Is_Tree(self.x-1,self.y)and not Is_Tree(self.x-1,self.y-1)):
                        self.move(0,-1)
                    elif (Is_Occupied_Actor(self.x-1,self.y)):
                        self.move(0,-1)
                    else:
                        self.move(-1,0)
                        
                elif(self.x < fruitX):
                    #pass on left side
                    if (Is_Tree(self.x+1,self.y+1)and not (Is_Tree(self.x,self.y+1))):
                        self.move(0,1)
                    elif (Is_Tree(self.x+1,self.y)and not Is_Tree(self.x+1,self.y+1)):
                        self.move(0,1)
                    else:
                        self.move(1,0)
                        
                elif(self.y < fruitY):
                    if (Is_Tree(self.x,self.y+1)):
                        self.move(-1,0)
                    elif (Is_Occupied_Actor(self.x,self.y+1)):
                        self.move(-1,0)
                    else:
                        self.move(0,1)
                        
                #(self.y > fruitY)       
                else:
                    if (Is_Tree(self.x,self.y-1)):
                        self.move(1,0)
                    else:
                        self.move(0,-1)    
    
        #kill orcs
        if (self.priority == 'killOrc'):
            #check that there exists orc in play
            existsOrc = False
            for actor in GAME_ACTORS:
                if (isinstance(actor,Orc)and not(actor.x==-1)):
                    existsOrc = True
            #if no orcs, wander
            if (existsOrc == False):
                
                randint = random.randint(1,5)
                #no move
                if (randint ==1):
                    self.move(0,0)
                #move left
                if randint == 2:
                    self.move(-1,0)
                #move right
                if randint==3:
                    self.move(1,0)
                #move down
                if randint == 4:
                    self.move(0,1)
                #move up
                if randint==5:
                    self.move(0,-1)
                
            if (existsOrc == True):
                #get x and y of nearest gold
                orcDistance = 1000
                orcX = 1000
                orcY = 1000
                for actor in GAME_ACTORS:
                    #get actors that are orcs and not out of play
                    if (isinstance(actor, Orc) and not(actor.x==-1)):
                        distance = abs(self.x-actor.x)+abs(self.y-actor.y)
                        if (distance < orcDistance):
                            orcDistance = distance
                            orcX = actor.x
                            orcY = actor.y

                #move towards closest gold and navigate around trees
                if (self.x > orcX):
                    #pass on right side
                    if (Is_Tree(self.x-1,self.y-1)and not (Is_Tree(self.x,self.y-1))):
                        self.move(0,-1)
                    elif (Is_Tree(self.x-1,self.y)and not Is_Tree(self.x-1,self.y-1)):
                        self.move(0,-1)
                    else:
                        self.move(-1,0)
                        
                elif(self.x < orcX):
                    #pass on left side
                    if (Is_Tree(self.x+1,self.y+1)and not (Is_Tree(self.x,self.y+1))):
                        self.move(0,1)
                    elif (Is_Tree(self.x+1,self.y)and not Is_Tree(self.x+1,self.y+1)):
                        self.move(0,1)
                    else:
                        self.move(1,0)
                        
                elif(self.y < orcY):
                    if (Is_Tree(self.x,self.y+1)):
                        self.move(-1,0)
                    else:
                        self.move(0,1)
                        
                #(self.y > orcY)       
                else:
                    if (Is_Tree(self.x,self.y-1)):
                        self.move(1,0)
                    else:
                        self.move(0,-1)
```

```python
class Rogue(Actor):
    def __init__(self,x,y):
        Actor.__init__(self, x, y)
        self.png = ROGUE
        self.attack_power = 2
        self.stamina=128
        self.name = "Rogue"
     
    def ply(self):
    
        #priority is initially set to gold each ply
        self.priority = 'gold'
        #if stamina is less than half, that becomes priority
        if (self.stamina <=64):
            self.priority = 'stamina'
        #if hp is less than half, that becomes priority
        if (self.hp<=4):
            self.priority = 'hp'
        
    
        #regain hp
        if (self.priority == 'hp'):
            #check that there exists Heart in play
            existsHeart = False
            for item in GAME_ITEMS:
                if (isinstance(item,Heart)and not (item.x==-1)):
                    existsHeart = True
            #if no heart in play, go find fruit
            if (existsHeart == False):
                self.priority = 'stamina'
                
            if (existsHeart == True):
                #get x and y of nearest Heart
                heartDistance = 1000
                heartX = 1000
                heartY = 1000
                for item in GAME_ITEMS:
                    #get items that are Heart and not out of play
                    if (isinstance(item, Heart) and not(item.x==-1)):
                        distance = abs(self.x-item.x)+abs(self.y-item.y)
                        if (distance < heartDistance):
                            heartDistance = distance
                            heartX = item.x
                            heartY = item.y

                #move towards closest heart and navigate around trees
                if (self.x > heartX):
                    #pass on right side
                    if (Is_Tree(self.x-1,self.y-1)and not (Is_Tree(self.x,self.y-1))):
                        self.move(0,-1)
                    elif (Is_Tree(self.x-1,self.y)and not Is_Tree(self.x-1,self.y-1)):
                        self.move(0,-1)
                    else:
                        self.move(-1,0)
                        
                elif(self.x < heartX):
                    #pass on left side
                    if (Is_Tree(self.x+1,self.y+1)and not (Is_Tree(self.x,self.y+1))):
                        self.move(0,1)
                    elif (Is_Tree(self.x+1,self.y)and not Is_Tree(self.x+1,self.y+1)):
                        self.move(0,1)
                    else:
                        self.move(1,0)
                        
                elif(self.y < heartY):
                    if (Is_Tree(self.x,self.y+1)):
                        self.move(-1,0)
                    else:
                        self.move(0,1)
                        
                #(self.y > heartY)       
                else:
                    if (Is_Tree(self.x,self.y-1)):
                        self.move(1,0)
                    else:
                        self.move(0,-1)    
    
        #regain stamina      
        if (self.priority == 'stamina'):

            #check that there exists Fruit in play
            existsFruit = False
            for item in GAME_ITEMS:
                if (isinstance(item,Fruit) and not (item.x==-1)):
                    existsFruit = True
            #if no fruit, find gold
            if (existsFruit == False):
                self.priority = 'gold'
            if (existsFruit == True):
                #get x and y of nearest fruit
                fruitDistance = 1000
                fruitX = 1000
                fruitY = 1000
                for item in GAME_ITEMS:
                    #get items that are fruit and not out of play
                    if (isinstance(item, Fruit) and not(item.x==-1)):
                        distance = abs(self.x-item.x)+abs(self.y-item.y)
                        if (distance < fruitDistance):
                            fruitDistance = distance
                            fruitX = item.x
                            fruitY = item.y
                #move towards closest fruit and navigate around trees
                if (self.x > fruitX):
                    #pass on right side
                    if (Is_Tree(self.x-1,self.y-1)and not (Is_Tree(self.x,self.y-1))):
                        self.move(0,-1)
                    elif (Is_Tree(self.x-1,self.y)and not Is_Tree(self.x-1,self.y-1)):
                        self.move(0,-1)
                    else:
                        self.move(-1,0)
                        
                elif(self.x < fruitX):
                    #pass on left side
                    if (Is_Tree(self.x+1,self.y+1)and not (Is_Tree(self.x,self.y+1))):
                        self.move(0,1)
                    elif (Is_Tree(self.x+1,self.y)and not Is_Tree(self.x+1,self.y+1)):
                        self.move(0,1)
                    else:
                        self.move(1,0)
                        
                elif(self.y < fruitY):
                    if (Is_Tree(self.x,self.y+1)):
                        self.move(-1,0)
                    else:
                        self.move(0,1)
                        
                #(self.y > fruitY)       
                else:
                    if (Is_Tree(self.x,self.y-1)):
                        self.move(1,0)
                    else:
                        self.move(0,-1)    
    
        #collect gold
        if (self.priority == 'gold'):
            #check that there exists Gold in play
            existsGold = False
            for item in GAME_ITEMS:
                if (isinstance(item,Gold)and not(item.x==-1)):
                    existsGold = True
            #if no gold, wander
            if (existsGold == False):
                
                randint = random.randint(1,5)
                #no move
                if (randint ==1):
                    self.move(0,0)
                #move left
                if randint == 2:
                    self.move(-1,0)
                #move right
                if randint==3:
                    self.move(1,0)
                #move down
                if randint == 4:
                    self.move(0,1)
                #move up
                if randint==5:
                    self.move(0,-1)
                
            if (existsGold == True):
                #get x and y of nearest gold
                goldDistance = 1000
                goldX = 1000
                goldY = 1000
                for item in GAME_ITEMS:
                    #get items that are gold and not out of play
                    if (isinstance(item, Gold) and not(item.x==-1)):
                        distance = abs(self.x-item.x)+abs(self.y-item.y)
                        if (distance < goldDistance):
                            goldDistance = distance
                            goldX = item.x
                            goldY = item.y

                #move towards closest gold and navigate around trees
                if (self.x > goldX):
                    #pass on right side
                    if (Is_Tree(self.x-1,self.y-1)and not (Is_Tree(self.x,self.y-1))):
                        self.move(0,-1)
                    elif (Is_Tree(self.x-1,self.y)and not Is_Tree(self.x-1,self.y-1)):
                        self.move(0,-1)
                    else:
                        self.move(-1,0)
                        
                elif(self.x < goldX):
                    #pass on left side
                    if (Is_Tree(self.x+1,self.y+1)and not (Is_Tree(self.x,self.y+1))):
                        self.move(0,1)
                    elif (Is_Tree(self.x+1,self.y)and not Is_Tree(self.x+1,self.y+1)):
                        self.move(0,1)
                    else:
                        self.move(1,0)
                        
                elif(self.y < goldY):
                    if (Is_Tree(self.x,self.y+1)):
                        self.move(-1,0)
                    else:
                        self.move(0,1)
                        
                #(self.y > goldY)       
                else:
                    if (Is_Tree(self.x,self.y-1)):
                        self.move(1,0)
                    else:
                        self.move(0,-1)
                
        return
```

### Helper Functions

```python
'''
returns whether or not a tile is occupied by
a game object
'''
def Is_Occupied_Actor(x,y):
    for actor in GAME_ACTORS:
        if (actor.x == x and actor.y==y):
            return True

'''
returns whether or not there is a tree
'''
def Is_Tree(x,y):
    if (GAME_MAP[x][y].blocks_path == True):
        return True   

    
'''
returns whether or not there is an item
'''    
def Is_Occupied_Item(x,y):
    for item in GAME_ITEMS:
        if (item.x ==x and item.y==y):
            return True
    
'''
function for initial generation of all of the game actors
'''
def Generate_Actors():
    #global vars
    global knight,rogue,orc1,orc2,orc3
    
    #randomely place the knight actor
    x=0
    y=0
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    knight = Knight(x,y)
    GAME_ACTORS.append(knight)

    #randomely place the rogue actor
    x=0
    y=0
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    rogue = Rogue(x,y)
    GAME_ACTORS.append(rogue)
    
    #randomly place orc1
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    orc1 =Orc(x,y)
    GAME_ACTORS.append(orc1)

    #randomly place orc2
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    orc2 =Orc(x,y)
    GAME_ACTORS.append(orc2)

    #generate orc3 out of play
    orc3 =Orc(-1,-1)
    GAME_ACTORS.append(orc3)

'''
function for initial generation of all of the game items
'''
def Generate_Items():
    
    #randomely place gold1
    x=0
    y=0
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Occupied_Item(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    gold1 = Gold(x,y)
    GAME_ITEMS.append(gold1)
    #randomely place gold2
    x=0
    y=0
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Occupied_Item(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    gold2 = Gold(x,y)
    GAME_ITEMS.append(gold2) 
    
    #generate gold3 out of play
    gold3 = Gold(-1,-1)
    GAME_ITEMS.append(gold3)
      
    #randomely place fruit1
    x=0
    y=0
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Occupied_Item(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    fruit1 = Fruit(x,y)
    GAME_ITEMS.append(fruit1)
    #randomely place fruit2
    x=0
    y=0
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Occupied_Item(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    fruit2 = Fruit(x,y)
    GAME_ITEMS.append(fruit2)
    
    #generate fruit3 out of play
    fruit3 = Fruit(-1,-1)
    GAME_ITEMS.append(fruit3)
    
    #randomely place heart1
    x=0
    y=0
    goodCoords = False
    while not(goodCoords):
        x = random.randint(1,MAP_WIDTH-1)
        y = random.randint(1,MAP_HEIGHT-1)
        if(not(Is_Occupied_Actor(x,y))and not(Is_Occupied_Item(x,y))and not(Is_Tree(x,y))):
            goodCoords = True
    heart1 = Heart(x,y)
    GAME_ITEMS.append(heart1)
    
    #generate heart2 out of play
    heart2 = Heart(-1,-1)
    GAME_ITEMS.append(heart2)
```

### Map

```python
'''
Starts by generating a field of grass and
then adds impassable trees around border
to keep agents from leaving. Then adds trees spread
around field for effect,
'''
def Create_Map():
    #map has a matrix representation
    #initialize as grassy tiles
    map = [[Tile(False,GRASSY_TILE) for y in range(0,MAP_HEIGHT)]for x in range(0,MAP_WIDTH)]

    #Set boundaries at edge of world so that agents cannot leave
    #and change tiles to tree
    for x in range (MAP_WIDTH):     
        map[x][0].blocks_path = True
        map[x][0].png = TREE
        map[x][MAP_HEIGHT-1].blocks_path = True
        map[x][MAP_HEIGHT-1].png = TREE
    for y in range(MAP_HEIGHT):
        map[0][y].blocks_path = True
        map[0][y].png = TREE
        map[MAP_WIDTH-1][y].blocks_path = True
        map[MAP_WIDTH-1][y].png = TREE
        
    #place some trees in middle of map
    map[7][9].blocks_path = True
    map[7][9].png = TREE
    map[11][4].blocks_path = True
    map[11][4].png = TREE
    map[13][17].blocks_path = True
    map[13][17].png = TREE
    map[19][2].blocks_path = True
    map[19][2].png = TREE
    map[16][12].blocks_path = True
    map[16][12].png = TREE
    map[4][12].blocks_path = True
    map[4][12].png = TREE
    map[3][16].blocks_path = True
    map[3][16].png = TREE
    map[24][8].blocks_path = True
    map[24][8].png = TREE
    map[26][15].blocks_path = True
    map[26][15].png = TREE

    return map
```

### Drawing

```python
def Draw_Game():
    
    #clear the surface
    MAIN_SURFACE.fill(BLACK_COLOR)
    
    #draw the game map
    Draw_Map(GAME_MAP)
    
    #draw all game items
    for item in GAME_ITEMS:
        item.draw()
    
    #draw all game actors
    for actor in GAME_ACTORS:
        actor.draw()
        
    #draw messages
    Draw_Messages()
    
    #update display
    pygame.display.flip()
      
    return

def Draw_Map(map): 
    #iterate through matrix and draw tiles
    for x in range(0,MAP_WIDTH):
        for y in range(0,MAP_HEIGHT):
            MAIN_SURFACE.blit(map[x][y].png,(x*CELL_WIDTH,y*CELL_HEIGHT))
    return

def Draw_Messages():
    #if num messages is less than max messages print all present
    if len(GAME_MESSAGES) <= MAX_MESSAGES: 
        messages_to_draw = GAME_MESSAGES
    #else, only print the max number of messages
    else:
        messages_to_draw = GAME_MESSAGES[-MAX_MESSAGES:]
    #get height of given font in pixels          
    font_height = Get_Font_Height(FONT_MESSAGE_TEXT)
    #y coordinate to start printing messages from; 
    y_begin = MAP_HEIGHT*CELL_HEIGHT - (MAX_MESSAGES * font_height)
    
    message_num = 0
    for message in messages_to_draw:
        Draw_Text(MAIN_SURFACE, message, (480, y_begin+(message_num*font_height)-3), WHITE_COLOR)
        message_num +=1   
        
def Draw_Text(surface, text, coordinates_tuple, text_color):
    #render the provided text in the proided gont
    text_surface = FONT_MESSAGE_TEXT.render(text, False, text_color) 
    #get the rectangle of the rendered object
    text_rectangle = text_surface.get_rect()
    #place the object
    text_rectangle.topleft = coordinates_tuple
    #blit it
    surface.blit(text_surface, text_rectangle)   
    
def Get_Font_Height(font):
    #render the surface with the provided font
    surface = font.render('a', False, (0,0,0))
    #get the rectangle for the surface
    rectangle = surface.get_rect()
    #return height
    return rectangle.height   
```

### Initialize the Game

```python
def Initialize_Game():
    '''
    initializes the game   
    '''
    
    #global vars
    global MAIN_SURFACE,GAME_MAP,GAME_MESSAGES,FONT_MESSAGE_TEXT,GAME_ITEMS,GAME_ACTORS
        
    
    
    #initialize pygame
    pygame.init()
    
    '''
    This font is made available for dekstop use by Typodermic Fonts at https://www.dafont.com/joystix.font
    '''
    #initialize font
    #https://www.dafont.com/joystix.font
    FONT_MESSAGE_TEXT = pygame.font.Font("C:\\Users\\gray\\Desktop\\AI\\Game\\PyWorld\\font\\joystix_monospace.ttf",12)
 
    #initialize the main surface as GAME_WIDTH x GAME_HEIGHT pixels
    MAIN_SURFACE = pygame.display.set_mode((MAP_WIDTH*CELL_WIDTH+CELL_WIDTH*25,MAP_HEIGHT*CELL_HEIGHT))

    #generate the game map
    GAME_MAP = Create_Map()
    
    #a list to store game messages in
    GAME_MESSAGES = []    
    
    #list of game actors that will be given a ply
    GAME_ACTORS = []
    #list of game items
    GAME_ITEMS = []    
    
    #generate the initial game actors
    Generate_Actors()

    #generate the initial game items
    Generate_Items()
    
    #draw the current state of the game
    Draw_Game()
    
    return
```

### Game Loop

```python
def Game_Loop():
    '''
    runs game loop until the game is over  
    '''
    
    #variable for terminating the game loop
    game_over = False
    quit = False
    
    #var for number of game loops
    loopCounter = 1
    
    #run game loop until the game is over
    while not(game_over):
        
        #get input as list of input (keyboard and mouse) events
        event_list = pygame.event.get()
        #process input event list
        for event in event_list:
            #event = 'click X button' -> quit game
            if event.type == pygame.QUIT:
                quit = True
                game_over = True
        #slow things down a little bit
        timer =10000000
        while not (timer <=0):
            timer-=1
        
        
        '''
        #actors move on space bar
        wait = True
        while(wait):
            #get input as list of input (keyboard and mouse) events
            event_list = pygame.event.get()
            #process input event list
            for event in event_list:
                #event = 'click X button' -> quit game
                if event.type == pygame.QUIT:
                    quit = True
                    game_over = True
                    wait = False
                if event.type == pygame.KEYDOWN:
                        wait = False
        '''
           
        #if an actor is not dead, they get a turn
        for actor in GAME_ACTORS:
            if not(actor.x==-1 and actor.y==-1):
                actor.ply()
        
          
        #if both knight and rogue are dead, break game loop. i.e. GAME_OVER
        if ((knight.x==-1 and knight.y==-1) and (rogue.x==-1 and rogue.y==-1)):
            game_over = True 
           
        #regenerate random discarded item every 20 game loops and place randomly
        if (loopCounter%10==0):
            RegenerateItem()
            
        #regenerate a dead orc every 30 loops and place randomly
        if (loopCounter%15==0):
            RegenerateOrc()
                       
        #draw the current state of the game
        Draw_Game()
        
        #increment loop counter
        loopCounter+=1
      
    if quit == True:
        pygame.quit()
        return
    else:
        Game_Over()

    return
```

### Game Over Screen

```python
'''
Displays the game over screen and waits for 
user to close the pygame window
'''
def Game_Over():
    #draw game over screen
    #clear the surface
    MAIN_SURFACE.fill(BLACK_COLOR)
    #draw a game over message
    Draw_Text(MAIN_SURFACE,"Game Over", (0,0), WHITE_COLOR)
    #update display
    pygame.display.flip()
    #wait for user to close the window
    close = False
    while not(close):
        event_list = pygame.event.get()
        for event in event_list:
            #event = 'click X button' -> quit game
            if event.type == pygame.QUIT:
                close = True
    
    pygame.quit()
    return
```

### Regenerate Items and Orcs

```python
def RegenerateItem():
    #check to make sure there is a discarded
    existsItem = False
    for item in GAME_ITEMS:
        if (item.x==-1 and item.y==-1):
            existsItem = True
    goodSelect = False
    if(existsItem):
        #pull random item, but make sure it is not on the map
        randItem =random.choice(GAME_ITEMS)
        while not(goodSelect):
            randItem = random.choice(GAME_ITEMS)
            if (randItem.x==-1 and randItem.y==-1):
                goodSelect=True
        #randomly place item
        x=0
        y=0
        goodCoords = False
        while not(goodCoords):
            x = random.randint(1,MAP_WIDTH-1)
            y = random.randint(1,MAP_HEIGHT-1)
            if(not(Is_Occupied_Actor(x,y))and not(Is_Occupied_Item(x,y))and not(Is_Tree(x,y))):
                goodCoords = True
        randItem.x = x
        randItem.y=y    
    return

def RegenerateOrc():
    #check to make sure there is a dead orc
    existsDeadOrc = False
    for actor in GAME_ACTORS:
        if (isinstance(actor,Orc) and actor.x==-1 and actor.y==-1):
            existsDeadOrc = True
    goodSelect = False
    if(existsDeadOrc):
        #pull random actor, but make sure it is an orc and not on the map
        randActor =random.choice(GAME_ACTORS)
        while not(goodSelect):
            randActor = random.choice(GAME_ACTORS)
            if (randActor.x==-1 and randActor.y==-1 and isinstance(randActor,Orc)):
                goodSelect=True
        #regenerate orc and randomly place
        randActor.hp=8
        randActor.stamina = 128
        x=0
        y=0
        goodCoords = False
        while not(goodCoords):
            x = random.randint(1,MAP_WIDTH-1)
            y = random.randint(1,MAP_HEIGHT-1)
            if(not(Is_Occupied_Actor(x,y))and not(Is_Occupied_Item(x,y))and not(Is_Tree(x,y))):
                goodCoords = True
        randActor.x = x
        randActor.y=y
    return
```
